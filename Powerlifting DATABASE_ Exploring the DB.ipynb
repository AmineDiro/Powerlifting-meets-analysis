```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('openpowerlifting.csv',low_memory=False)
```

# Some data exploration

```python
df_tested = df[df['Tested']=='Yes']
df_Ntested = df[df['Tested']!='Yes']
df_tested=df_tested[['Sex', 'Age','BodyweightKg','TotalKg','Place','Best3SquatKg','Best3BenchKg','Best3DeadliftKg']]
```

```python
sns.countplot(x='Sex',data=df_tested)
```

```python
df_tested['AgeCate']=df[df['Tested']=='Yes']['AgeClass']
```

```python
s=df_tested[df_tested['AgeCate'].isin(['20-23', '24-34','35-39'])]
sns.violinplot(x='AgeCate',y='TotalKg',data=s,hue='Sex',split=True,palette='coolwarm')
```

```python
#sns.countplot(df['Equipment'])

# Pie chart
labels = df['Equipment'].value_counts().index.tolist()
sizes = (df['Equipment'].value_counts().values)/len(df)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
